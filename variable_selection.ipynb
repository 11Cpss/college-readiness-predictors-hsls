# Making New CSV File with targetted Variables
### Condensed old variables, and added the new ones that was analyzed above
- Combined sat scores.   
- combined ACT scores.    
- combined hours worked on homework.   
- combined total ap scores

```python
import pandas as pd
import numpy as np

df = pd.read_csv('hsls_17_student_pets_sr_v1_0.csv')
original_vars = [
    "X3TNEWBASIC",
    "X3TGPA12TH",
    "X3TGPA11TH",
    "X3TGPA10TH",
    "X3TGPA9TH",
    "X3TGPAACAD",
    "X3TGPAWGT",
    "C2AVGSATMATH",
    "C2AVGSATREAD",
    "C2AVGSATWRIT",
    "C2AVGACTCOMP",
    "S1HRMHOMEWK",
    "S1HRSHOMEWK",
    "S1HROTHHOMWK",
    "X5YR1GPA",
    "X5YR1ATT",
    "X5YR1ERN",
    "X5REMRAT",
    "X5ENRATT",
    "X1FAMINCOME",
    "X1HHNUMBER",
    "X1SES_U",
    "X3NUMHSATTND",
    "X2PROBLEM",
    "X2SCHOOLCLI",
    "X2BEHAVEIN",
    'X3TCREDAPIB'
]

df = df[original_vars]
```

```python
columns_remame = {
    'X3TNEWBASIC':'RIGOR',
    'X3TGPA12TH': 'HS_GPA_12TH',
    'X3TGPA11TH': 'HS_GPA_11TH',
    'X3TGPA10TH': 'HS_GPA_10TH',
    'X3TGPA9TH': 'HS_GPA_9TH',
    'X3TGPAACAD': 'HS_GPA_ACAD',
    'X3TGPAWGT': 'HS_GPA_WGT',
    'C2AVGSATMATH': 'SAT_MATH_AVG',
    'C2AVGSATREAD': 'SAT_READ_AVG',
    'C2AVGSATWRIT': 'SAT_WRIT_AVG',
    'C2AVGACTCOMP': 'ACT_SCHOOL_AVG',
    'S1HRMHOMEWK': 'HOUR_MATH_HOMEWORK',
    'S1HRSHOMEWK': 'HOUR_SCIENCE_HOMEWORK',
    'S1HROTHHOMWK': 'HOUR_OTHER_HOMEWORK',
    'X3TCREDAPIB': 'TOTAL_AP_CREDITS',
    'X5YR1GPA': 'FRESHMAN_GPA',
    'X5YR1ATT': 'FRESHMAN_CREDITS_ATTEMPTED',
    'X5YR1ERN': 'FRESHMAN_CREDITS_EARNED',
    "X1FAMINCOME": "FAMILY_INCOME_BRACKET",
    "X1HHNUMBER": "HOUSEHOLD_SIZE",
    "X1SES_U": "SES_COMPOSITE_URBANICITY_ADJ",
    "X3NUMHSATTND": "NUM_HS_ATTENDED",
    "X2PROBLEM": "SCHOOL_PROBLEM_SCALE",
    "X2SCHOOLCLI": "SCHOOL_CLIMATE_SCALE",
    "X2BEHAVEIN": "STUDENT_SCHOOL_MOTIVATION",
    "X5REMRAT": "REMEDIAL_COURSE_RATIO",
    "X5ENRATT": "ENROLLMENT_ATTAINMENT_STATUS",
  

}
df = df.rename(columns=columns_remame)
```

```python
df['SAT_SCHOOL_AVG'] = df['SAT_MATH_AVG'] + df['SAT_READ_AVG']
df['HOURS_HOMEWORK'] = df['HOUR_MATH_HOMEWORK'] + df['HOUR_SCIENCE_HOMEWORK'] + df['HOUR_OTHER_HOMEWORK']

columns_drop = ['SAT_MATH_AVG','SAT_READ_AVG', 
                'HOUR_MATH_HOMEWORK', 'HOUR_SCIENCE_HOMEWORK', 'HOUR_OTHER_HOMEWORK']
df = df.drop(columns=columns_drop)
```
